```python
%gui qt
import vtk
import numpy as np
import mayavi
from mayavi import mlab
import os

# Mayavi config
mlab.options.backend = 'envisage'

# Input and output dirs
input_dir = '/Users/itxaso/codiTFG/ChiralStructuresProject/ChiralStructures25/out'
output_dir = '/Users/itxaso/codiTFG/ChiralStructuresProject/ChiralStructures25/Images'

# Change entrance dir
os.chdir(input_dir)

# Filtrar archivos de configuración angular
conf_end_angle_list = [f for f in os.listdir() if f.startswith("Conf_end_N")]

def draw_cylinder():
    """Draws a cylindrical surface."""
    # Define cylinder parameters
    radius = 3.6013/2
    height = 10.606*2
#    radius = 5.40195/2
#    height = 10.606*2
#    radius = 3.6013
#    height = 10.606
    theta = np.linspace(0, 2*np.pi, 100)  # Angular coordinates
    z = np.linspace(-height/2, height/2, 50)  # Height coordinates

    # Create meshgrid for cylindrical coordinates
    Theta, Z = np.meshgrid(theta, z)
    X = radius * np.cos(Theta)
    Y = radius * np.sin(Theta)

    # Plot the surface of the cylinder
    cyl = mlab.mesh(X, Y, Z, color=(0.85, 0.9, 0.82),opacity=0.5, colormap='viridis')  # Light blue color
    cyl.actor.mapper.scalar_visibility = False
    return cyl

# Function for processing all files
def process_file(file_name):

    # Extract information
    points = np.loadtxt(file_name)
    x0, y0, z0 = points[:, 0], points[:, 1], points[:, 2]
    sx, sy, sz = points[:, 3], points[:, 4], points[:, 5]
    height = np.max(z0)
    print(height)

    # Create figure and configurate scene
    fig = mlab.figure(size=(500, 500), bgcolor=(0.82, 0.82, 0.82))
    res = mlab.quiver3d(x0, y0, z0, sx, sy, sz, scalars=sx, line_width=3, mode='sphere', scale_factor=0.8)
    # Assuming 'lut' is defined somewhere before this function call
    # If not, you might need to create it or use a default colormap
    try:
        res.module_manager.scalar_lut_manager.lut.table = lut
    except NameError:
        print("Warning: 'lut' is not defined. Using default colormap.")
    res.glyph.glyph_source.glyph_position = 'center'
    res.glyph.color_mode = 'color_by_scalar'


    # Color bar configuration
    res.module_manager.scalar_lut_manager.label_text_property.font_size = 8
    res.module_manager.scalar_lut_manager.label_text_property.italic = False
    res.module_manager.scalar_lut_manager.scalar_bar.title = 'Sz'
    res.module_manager.scalar_lut_manager.title_text_property.font_size = 7
    #res.glyph.glyph_source.glyph_source.tip_resolution = 24
    #res.glyph.glyph_source.glyph_source.tip_radius = 0.1
    #res.glyph.glyph_source.glyph_source.shaft_radius = 0.05

    # Adjust camera and lights
    try:
        engine = mlab.get_engine()
    except NameError:
        from mayavi.api import Engine
        engine = Engine()
        engine.start()
    if len(engine.scenes) == 0:
        engine.new_scene()

    scene = engine.scenes[0]
    scene.scene.isometric_view()  # Isometric view
    scene = fig.scene
    scene.light_manager.lights[1].intensity = 1.0
    scene.light_manager.lights[2].intensity = 1.0
    scene.render()

    # Draw the cylindrical surface
    draw_cylinder()

    # Save images
    output_file = os.path.join(output_dir, file_name.replace('.out', '.png'))
    mlab.savefig(output_file, size=(5000, 5000))
    print(f"Saved image: {output_file}")

    # Close scene
    #mlab.close(fig)


# Process all files inside the loop
for file in conf_end_angle_list:
    process_file(file)

# To just draw the cylinder without processing files, uncomment the following lines:
#mlab.figure(size=(500, 500), bgcolor=(0.82, 0.82, 0.82))
#draw_cylinder()
#mlab.show()
```

```python
%gui qt
import vtk
import numpy as np
import mayavi
from mayavi import mlab
import os

# Mayavi configuration
mlab.options.backend = 'envisage'

# Input and output dir
input_dir = '/Users/itxaso/codiTFG/ChiralStructuresProject/ChiralStructures25/out'
output_dir = '/Users/itxaso/codiTFG/ChiralStructuresProject/ChiralStructures25/Images'
os.chdir(input_dir)

# List of plane config
conf_end_angle_list = [f for f in os.listdir() if f.startswith("Conf_end_N")]

def process_file(file_name):
    points = np.loadtxt(file_name)    

    x0, y0, z0 = points[:, 0], points[:, 1], points[:, 2]
    sx, sy, sz = points[:, 3], points[:, 4], points[:, 5]
    height = np.max(z0)
    print(height)

    # Flat coordinates for Angle (transformation).
    r = np.sqrt(x0**2 + y0**2)
    xp = r * np.arctan2(y0, x0)  # new x coordinate
    yp = z0             # new y coordinate
    zp = np.zeros_like(xp)

    # Spin vector transformation
    sx2 = -sx * np.sin(np.arctan2(y0,x0)) + sy * np.cos(np.arctan2(y0,x0))
    sy2 = sz
    sz2 = sx * np.cos(np.arctan2(y0,x0)) + sy * np.sin(np.arctan2(y0,x0))
    
    # Figure
    fig = mlab.figure(size=(10000, 10000), bgcolor=(0.82, 0.82, 0.82))
    res = mlab.quiver3d(xp, yp, zp, sx2, sy2, sz2, scalars=sy2, line_width=3, mode='cone', scale_factor=0.8)
    res.module_manager.scalar_lut_manager.lut.table = lut
    res.glyph.glyph_source.glyph_position = 'center'
    res.glyph.color_mode = 'color_by_scalar'


    # View and lights
    try:
        engine = mlab.get_engine()
    except NameError:
        from mayavi.api import Engine
        engine = Engine()
        engine.start()
    if len(engine.scenes) == 0:
        engine.new_scene()
        
    scene = fig.scene
    scene.z_plus_view()  
    scene.light_manager.lights[1].intensity = 1.0
    scene.light_manager.lights[2].intensity = 1.0
    scene.render()

    # Save image
    output_file = os.path.join(output_dir, "Plane_" + file_name.replace('.out', '.png'))
    mlab.savefig(output_file, size=(5000, 5000))
    print(f"Imagen guardada: {output_file}")

    mlab.close()

# Loop for all files in the list
for file in conf_end_angle_list:
    process_file(file)
```

### **Summary of the Columns**
| Column | Description                                      |
|-------|--------------------------------------------------|
| 1     | X coordinate of the position \(R(i,1)\)           |
| 2     | Y coordinate of the position \(R(i,2)\)           |
| 3     | Z coordinate of the position \(R(i,3)\)           |
| 4     | X component of the spin \(Spin(i,1)\)              |
| 5     | Y component of the spin \(Spin(i,2)\)              |
| 6     | Z component of the spin \(Spin(i,3)\)              |
| 7    | Polar angle \(\Theta\)                             |
| 8    | Azimuthal angle \(\Phi\)                           |

#### **Flat Coordinates Transformation:**  

$r = \sqrt{x^2 + y^2}$

$\phi = \arctan2(y, x)$

$x' = r \cdot \phi$

$y' = z$
 

#### **Spin Vector Transformation:**  
$\phi_0 = \arctan2(y, x)$ 

$s_x' = -s_x \sin(\phi_0) + s_y \cos(\phi_0)$ 

$s_y' = s_z$ 

$s_z' = s_x \cos(\phi_0) + s_y \sin(\phi_0)$
